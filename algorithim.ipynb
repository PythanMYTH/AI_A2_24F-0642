```python
import heapq
import time

def reconstruct_path(current, draw):
    while current.parent:
        current = current.parent
        current.make_path()
        draw()


def greedy(draw, grid, start, goal, heuristic):
    count = 0
    open_set = []
    heapq.heappush(open_set, (0, count, start))
    visited = set()

    nodes_expanded = 0
    start_time = time.time()

    while open_set:
        current = heapq.heappop(open_set)[2]

        if current == goal:
            reconstruct_path(goal, draw)
            return True, nodes_expanded, (time.time() - start_time) * 1000

        visited.add(current)
        nodes_expanded += 1

        for neighbor in current.neighbors:
            if neighbor not in visited:
                count += 1
                h = heuristic(neighbor.get_pos(), goal.get_pos())
                heapq.heappush(open_set, (h, count, neighbor))
                neighbor.parent = current
                neighbor.make_open()

        draw()
        current.make_closed()

    return False, nodes_expanded, 0


def astar(draw, grid, start, goal, heuristic):
    count = 0
    open_set = []
    heapq.heappush(open_set, (0, count, start))

    start.g = 0
    start.f = heuristic(start.get_pos(), goal.get_pos())

    nodes_expanded = 0
    start_time = time.time()

    open_hash = {start}

    while open_set:
        current = heapq.heappop(open_set)[2]
        open_hash.remove(current)

        if current == goal:
            reconstruct_path(goal, draw)
            return True, nodes_expanded, (time.time() - start_time) * 1000

        nodes_expanded += 1

        for neighbor in current.neighbors:
            temp_g = current.g + 1

            if temp_g < neighbor.g:
                neighbor.parent = current
                neighbor.g = temp_g
                neighbor.h = heuristic(neighbor.get_pos(), goal.get_pos())
                neighbor.f = neighbor.g + neighbor.h

                if neighbor not in open_hash:
                    count += 1
                    heapq.heappush(open_set, (neighbor.f, count, neighbor))
                    open_hash.add(neighbor)
                    neighbor.make_open()

        draw()
        current.make_closed()

    return False, nodes_expanded, 0
```
