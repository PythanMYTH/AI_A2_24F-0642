```python
import pygame
from config import WIDTH, ROWS
from grid import make_grid, draw
from algorithms import astar, greedy
from heuristics import manhattan, euclidean
from dynamic import spawn_dynamic_obstacles

pygame.init()
WIN = pygame.display.set_mode((WIDTH, WIDTH))
pygame.display.set_caption("Dynamic Pathfinding Agent")

def main():
    grid = make_grid(ROWS, WIDTH)

    start = grid[0][0]
    goal = grid[ROWS-1][ROWS-1]

    start.make_start()
    goal.make_goal()

    run = True
    algorithm = "astar"
    heuristic = manhattan
    dynamic_mode = False

    while run:
        draw(WIN, grid, ROWS, WIDTH)

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                run = False

            if event.type == pygame.KEYDOWN:

                if event.key == pygame.K_a:
                    algorithm = "astar"

                if event.key == pygame.K_g:
                    algorithm = "greedy"

                if event.key == pygame.K_h:
                    heuristic = euclidean if heuristic == manhattan else manhattan

                if event.key == pygame.K_d:
                    dynamic_mode = not dynamic_mode

                if event.key == pygame.K_SPACE:
                    for row in grid:
                        for node in row:
                            node.update_neighbors(grid)

                    if algorithm == "astar":
                        found, nodes, time_taken = astar(
                            lambda: draw(WIN, grid, ROWS, WIDTH),
                            grid, start, goal, heuristic
                        )
                    else:
                        found, nodes, time_taken = greedy(
                            lambda: draw(WIN, grid, ROWS, WIDTH),
                            grid, start, goal, heuristic
                        )

                    print("Nodes Expanded:", nodes)
                    print("Execution Time (ms):", time_taken)

                if event.key == pygame.K_r:
                    grid = make_grid(ROWS, WIDTH)
                    start = grid[0][0]
                    goal = grid[ROWS-1][ROWS-1]
                    start.make_start()
                    goal.make_goal()

        if dynamic_mode:
            spawn_dynamic_obstacles(grid, start, goal)

    pygame.quit()

if __name__ == "__main__":
    main()
```
